### Linear Regression with 'acousticness' and 'energy'

Now, let's build a linear regression model to predict `popularity` using both `acousticness` and `energy` as independent variables. We will follow similar steps: prepare the data, split it into training and testing sets, train the model, and evaluate its performance.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df_cleaned_lr_multi = df.dropna(subset=['popularity', 'acousticness', 'energy']).copy()
X_multi = df_cleaned_lr_multi[['acousticness', 'energy']]
y_multi = df_cleaned_lr_multi['popularity']

X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X_multi, y_multi, test_size=0.2, random_state=42)

model_multi = LinearRegression()
model_multi.fit(X_train_multi, y_train_multi)

y_pred_multi = model_multi.predict(X_test_multi)

mse_multi = mean_squared_error(y_test_multi, y_pred_multi)
r2_multi = r2_score(y_test_multi, y_pred_multi)

print(f"Mean Squared Error (Multi-feature): {mse_multi:.2f}")
print(f"R-squared (Multi-feature): {r2_multi:.2f}")
print(f"Coefficients: {model_multi.coef_}")
print(f"Intercept: {model_multi.intercept_:.2f}")
```

### Visualization of Multi-feature Linear Regression

Visualizing a multi-feature linear regression can be challenging in 2D. However, we can visualize the effect of each feature on `popularity` while holding other features constant (e.g., at their mean value).

```python
import numpy as np

plot_df = X_test_multi.copy()
plot_df['popularity_actual'] = y_test_multi
plot_df['popularity_predicted'] = y_pred_multi

mean_acousticness = X_train_multi['acousticness'].mean()
mean_energy = X_train_multi['energy'].mean()

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(plot_df['acousticness'], plot_df['popularity_actual'], color='blue', alpha=0.5, label='Actual Data')

x_acousticness_plot = np.linspace(plot_df['acousticness'].min(), plot_df['acousticness'].max(), 100)
y_acousticness_pred_line = model_multi.intercept_ + model_multi.coef_[0] * x_acousticness_plot + model_multi.coef_[1] * mean_energy
plt.plot(x_acousticness_plot, y_acousticness_pred_line, color='red', linewidth=2, label='Regression Line (energy at mean)')

plt.xlabel('Acousticness')
plt.ylabel('Popularity')
plt.title('Linear Regression: Acousticness vs. Popularity')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(plot_df['energy'], plot_df['popularity_actual'], color='blue', alpha=0.5, label='Actual Data')

x_energy_plot = np.linspace(plot_df['energy'].min(), plot_df['energy'].max(), 100)
y_energy_pred_line = model_multi.intercept_ + model_multi.coef_[0] * mean_acousticness + model_multi.coef_[1] * x_energy_plot
plt.plot(x_energy_plot, y_energy_pred_line, color='red', linewidth=2, label='Regression Line (acousticness at mean)')

plt.xlabel('Energy')
plt.ylabel('Popularity')
plt.title('Linear Regression: Energy vs. Popularity')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

### Statistical Tests

#### 1. One-Sample Z-Test

The Z-test is used to compare a sample mean to a known population mean when the population standard deviation is known, or when the sample size is large (n > 30), allowing the sample standard deviation to approximate the population standard deviation. Here, we'll assume a hypothetical population mean for 'popularity' and test if our sample mean significantly differs from it.

```python
from statsmodels.stats.weightstats import ztest

df_ztest = df.dropna(subset=['popularity']).copy()

sample_popularity = df_ztest['popularity']

hypothesized_mean = 20

z_statistic, p_value = ztest(sample_popularity, value=hypothesized_mean)

print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print(f"\nConclusion: With a p-value of {p_value:.3f}, we reject the null hypothesis. The sample mean popularity is significantly different from the hypothesized population mean of {hypothesized_mean}.")
else:
    print(f"\nConclusion: With a p-value of {p_value:.3f}, we fail to reject the null hypothesis. There is no significant difference between the sample mean popularity and the hypothesized population mean of {hypothesized_mean}.")
```

#### 2. Independent Two-Sample T-Test

The independent two-sample T-test is used to determine if there is a significant difference between the means of two independent groups. Here, we'll compare the 'popularity' of songs based on their 'mode' (0 or 1, representing minor or major key, respectively).

```python
from scipy import stats

df_ttest = df.dropna(subset=['popularity', 'mode']).copy()

popularity_mode_0 = df_ttest[df_ttest['mode'] == 0]['popularity']
popularity_mode_1 = df_ttest[df_ttest['mode'] == 1]['popularity']

t_statistic, p_value = stats.ttest_ind(popularity_mode_0, popularity_mode_1, equal_var=False) # Welch's t-test assuming unequal variances

print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print(f"\nConclusion: With a p-value of {p_value:.3f}, we reject the null hypothesis. There is a significant difference in popularity between songs with mode 0 and mode 1.")
else:
    print(f"\nConclusion: With a p-value of {p_value:.3f}, we fail to reject the null hypothesis. There is no significant difference in popularity between songs with mode 0 and mode 1.")
```

#### 3. ANOVA (Analysis of Variance) Test

ANOVA is used to test for differences between three or more group means. Here, we'll examine if there's a significant difference in 'popularity' across different 'languages'.

```python
from scipy import stats

df_anova = df.dropna(subset=['popularity', 'language']).copy()

language_groups = [df_anova['popularity'][df_anova['language'] == lang] for lang in df_anova['language'].unique()]

language_groups = [group for group in language_groups if not group.empty]

f_statistic, p_value = stats.f_oneway(*language_groups)

print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print(f"\nConclusion: With a p-value of {p_value:.3f}, we reject the null hypothesis. There is a significant difference in popularity among songs of different languages.")
else:
    print(f"\nConclusion: With a p-value of {p_value:.3f}, we fail to reject the null hypothesis. There is no significant difference in popularity among songs of different languages.")
```

#### 4. Chi-Square Test of Independence

The Chi-square test of independence is used to determine if there is a significant association between two categorical variables. Here, we'll investigate if there's an association between 'language' and 'high_popularity'.

```python
from scipy import stats
import pandas as pd

median_popularity = df['popularity'].median()
df['high_popularity'] = (df['popularity'] > median_popularity).astype(int)

df_chi2 = df.dropna(subset=['language', 'high_popularity']).copy()

contingency_table = pd.crosstab(df_chi2['language'], df_chi2['high_popularity'])

print("Contingency Table (Language vs. High Popularity):\n", contingency_table)

chi2_statistic, p_value, dof, expected_freq = stats.chi2_contingency(contingency_table)

print(f"\nChi-square statistic: {chi2_statistic:.2f}")
print(f"P-value: {p_value:.3f}")
print(f"Degrees of Freedom: {dof}")

if p_value < 0.05:
    print(f"\nConclusion: With a p-value of {p_value:.3f}, we reject the null hypothesis. There is a significant association between language and high popularity.")
else:
    print(f"\nConclusion: With a p-value of {p_value:.3f}, we fail to reject the null hypothesis. There is no significant association between language and high popularity.")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

df_cleaned_poly = df.dropna(subset=['popularity', 'acousticness']).copy()

X_poly = df_cleaned_poly[['acousticness']]
y_poly = df_cleaned_poly['popularity']

X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(X_poly, y_poly, test_size=0.2, random_state=42)

poly_model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())

# Train the model
poly_model.fit(X_train_poly, y_train_poly)

# Make predictions on the test set
y_pred_poly = poly_model.predict(X_test_poly)

# Evaluate the model
mse_poly = mean_squared_error(y_test_poly, y_pred_poly)
r2_poly = r2_score(y_test_poly, y_pred_poly)

print(f"Mean Squared Error (Polynomial Regression): {mse_poly:.2f}")
print(f"R-squared (Polynomial Regression): {r2_poly:.2f}")

# To get coefficients, we can access the LinearRegression model within the pipeline
linear_reg_model = poly_model.named_steps['linearregression']
print(f"Polynomial Coefficients: {linear_reg_model.coef_}")
print(f"Polynomial Intercept: {linear_reg_model.intercept_:.2f}")
```

### Logistic Regression

Logistic Regression is used for binary classification tasks. Since the `popularity` column is continuous, we need to convert it into a binary target variable. I will create a new column, `high_popularity`, where songs with popularity above the median will be labeled as `1` (high popularity) and others as `0` (low popularity).

```python
median_popularity = df['popularity'].median()
df['high_popularity'] = (df['popularity'] > median_popularity).astype(int)

display(df[['popularity', 'high_popularity']].head())
display(df['high_popularity'].value_counts())
```

Now, let's implement the Logistic Regression model using `acousticness` and `energy` as features to predict `high_popularity`.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

df_cleaned_lr_logistic = df.dropna(subset=['high_popularity', 'acousticness', 'energy']).copy()

X_logistic = df_cleaned_lr_logistic[['acousticness', 'energy']]
y_logistic = df_cleaned_lr_logistic['high_popularity']

X_train_logistic, X_test_logistic, y_train_logistic, y_test_logistic = train_test_split(X_logistic, y_logistic, test_size=0.2, random_state=42)

model_logistic = LogisticRegression(random_state=42)
model_logistic.fit(X_train_logistic, y_train_logistic)

y_pred_logistic = model_logistic.predict(X_test_logistic)

accuracy = accuracy_score(y_test_logistic, y_pred_logistic)
conf_matrix = confusion_matrix(y_test_logistic, y_pred_logistic)
class_report = classification_report(y_test_logistic, y_pred_logistic)

print(f"Accuracy: {accuracy:.2f}")
print("\nConfusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)
print(f"Coefficients: {model_logistic.coef_}")
print(f"Intercept: {float(model_logistic.intercept_[0]):.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a meshgrid to plot the decision boundary
x_min, x_max = X_logistic['acousticness'].min() - 0.1, X_logistic['acousticness'].max() + 0.1
y_min, y_max = X_logistic['energy'].min() - 0.1, X_logistic['energy'].max() + 0.1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

# Predict on the meshgrid points
Z = model_logistic.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary and the data points
plt.figure(figsize=(10, 7))
plt.contourf(xx, yy, Z, alpha=0.4, cmap=plt.cm.RdBu)

# Plot the actual data points
plt.scatter(X_test_logistic['acousticness'], X_test_logistic['energy'], c=y_test_logistic, cmap=plt.cm.RdBu, edgecolors='k', s=20)
plt.xlabel('Acousticness')
plt.ylabel('Energy')
plt.title('Logistic Regression Decision Boundary for High Popularity')
plt.colorbar(label='High Popularity (0=No, 1=Yes)')
plt.grid(True)
plt.show()
```

### Making Predictions with Logistic Regression

Now that the Logistic Regression model (`model_logistic`) is trained, we can use it to predict the `high_popularity` of new songs based on their `acousticness` and `energy` values. The model outputs probabilities, which are then converted into binary classifications (0 or 1).

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

df_cleaned_lr_logistic = df.dropna(subset=['high_popularity', 'acousticness', 'energy']).copy()
X_logistic = df_cleaned_lr_logistic[['acousticness', 'energy']]
y_logistic = df_cleaned_lr_logistic['high_popularity']
X_train_logistic, X_test_logistic, y_train_logistic, y_test_logistic = train_test_split(X_logistic, y_logistic, test_size=0.2, random_state=42)
model_logistic = LogisticRegression(random_state=42)
model_logistic.fit(X_train_logistic, y_train_logistic)


new_songs_data = pd.DataFrame({
    'acousticness': [0.1, 0.8, 0.3, 0.9, 0.05],
    'energy': [0.9, 0.2, 0.7, 0.1, 0.85]
})


probabilities = model_logistic.predict_proba(new_songs_data)

predictions = model_logistic.predict(new_songs_data)

new_songs_data['low_popularity_probability'] = probabilities[:, 0]
new_songs_data['high_popularity_probability'] = probabilities[:, 1]
new_songs_data['predicted_high_popularity'] = predictions

display(new_songs_data)
```

In the output above:
- `low_popularity_probability`: The model's estimated probability that the song has low popularity.
- `high_popularity_probability`: The model's estimated probability that the song has high popularity.
- `predicted_high_popularity`: The final binary prediction (0 for low popularity, 1 for high popularity), typically based on a 0.5 probability threshold.
